```python
# finding price outliers

import numpy as np

# Filtering rows where price = 0
filtered_cars_df = used_cars_df[used_cars_df['price'] != 0]

# Finding averages
car_prices = filtered_cars_df["price"].values
average_price = np.percentile(car_prices, 50)
print(f'Average price: {average_price}')
q3_price = np.percentile(car_prices, 75)
print(f'Upper quartile: {q3_price}')
q1_price = np.percentile(car_prices, 25)
print(f'Lower quartile: {q1_price}')
iqr = q3_price - q1_price
print(f'Interquartile range: {iqr}')

# Finding outliers
lower_outlier_value = q1_price - 1.5 * iqr
print(f'Lower outlier critical value: {lower_outlier_value}')
upper_outlier_value = q3_price + 1.5 * iqr
print(f'Upper outlier critical value: {upper_outlier_value}')

# Creating dataframe without outliers
clean_cars_df = filtered_cars_df[filtered_cars_df['price'] < 58475.0]
clean_car_prices = clean_cars_df["price"]
print(f'minimum car price: {clean_car_prices.min()}')
```

Data analysis and visualisation:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.style as style 
from matplotlib import *
style.use('fivethirtyeight')
```

Investigating average price per year:

```python
# Comparing year to price

plt.scatter(clean_cars_df["year"].values, clean_cars_df["price"].values, color = 'orange', marker = 'x', s = 10)
plt.title('Car prices per year')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()
```

```python
# Creating graph to show average price per year

avg_price_per_year = clean_cars_df.groupby('year')['price'].mean()

# Plotting
avg_price_per_year.plot(kind='line', marker='o', title='Average Price per Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

```python
# showing modern trends of price by year

modern_used_cars_df = clean_cars_df[clean_cars_df['year'] >= 2000.0]
modern_avg_price_per_year = modern_used_cars_df.groupby('year')['price'].mean()
modern_avg_price_per_year.plot(kind='line', marker='o', title='Average Price per Year', color = 'green')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

```python
# showing vintage trends of price by year

vintage_used_cars_df = clean_cars_df[clean_cars_df['year'] < 2000.0]
vintage_avg_price_per_year = vintage_used_cars_df.groupby('year')['price'].mean()
vintage_avg_price_per_year.plot(kind='line', marker='o', title='Average Price per Year', color = 'red')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.show()
```

Insights:

    - The price of used car decreases from 1910 to 2000.
    - The price of used cars increases from 2000 to 2020.

Investigating average price by manufacturer:

```python
# comparing average price by manufacturer

avg_manufacturer_price = clean_cars_df.groupby('manufacturer')['price'].mean().reset_index(name = 'Average price')
avg_manufacturer_price = avg_manufacturer_price.sort_values('Average price', ascending=True)
plt.bar(avg_manufacturer_price['manufacturer'], avg_manufacturer_price['Average price'])
plt.title('Average used price by manufacturer')
plt.ylabel('Average price')
plt.xlabel('Manufacturer')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
print(clean_cars_df['condition'].unique())
```

```python
# Finding reasons for average prices of manufacturers, starting with percentage in new condition per manufacturer

manufacturer_condition_new = clean_cars_df.groupby('manufacturer')['condition'].apply(lambda x: ((x == 'new').sum() / (x != '').sum()) * 100).reset_index(name='New percentage')
manufacturer_condition_new = manufacturer_condition_new.sort_values('New percentage', ascending=True)
plt.barh(y = manufacturer_condition_new['manufacturer'].values, width = manufacturer_condition_new['New percentage'].values, color = 'brown')
plt.rcParams['figure.figsize'] = [8, 8]
plt.title('Percentage of new cars per manufacturer')
plt.xlabel('Percentage of new cars')
plt.ylabel('Manufacturer')
plt.show()
```

```python
manufacturer_condition_like_new = clean_cars_df.groupby('manufacturer')['condition'].apply(lambda x: ((x == 'like new').sum() / (x != '').sum()) * 100).reset_index(name='Like new percentage')
manufacturer_condition_like_new = manufacturer_condition_like_new.sort_values('Like new percentage', ascending=True)
plt.barh(y = manufacturer_condition_like_new['manufacturer'].values, width = manufacturer_condition_like_new['Like new percentage'].values, color = 'brown')
plt.title('Percentage of like-new cars per manufacturer')
plt.xlabel('Percentage of like-new cars')
plt.ylabel('Manufacturer')
plt.show()
```

```python
manufacturer_condition_excellent = clean_cars_df.groupby('manufacturer')['condition'].apply(lambda x: ((x == 'excellent').sum() / (x != '').sum()) * 100).reset_index(name='New percentage')
manufacturer_condition_excellent = manufacturer_condition_excellent.sort_values('New percentage', ascending=True)
plt.barh(y = manufacturer_condition_excellent['manufacturer'].values, width = manufacturer_condition_excellent['New percentage'].values, color = 'brown')
plt.title('Percentage of excellent condition cars per manufacturer')
plt.xlabel('Percentage of excellent condition cars')
plt.ylabel('Manufacturer')
plt.show()
```

```python
manufacturer_condition_good = clean_cars_df.groupby('manufacturer')['condition'].apply(lambda x: ((x == 'good').sum() / (x != '').sum()) * 100).reset_index(name='Good percentage')
manufacturer_condition_good = manufacturer_condition_good.sort_values('Good percentage', ascending=True)
plt.barh(y = manufacturer_condition_good['manufacturer'].values, width = manufacturer_condition_good['Good percentage'].values, color = 'brown')
plt.title('Percentage of good condition cars per manufacturer')
plt.xlabel('Percentage of good condition cars')
plt.ylabel('Manufacturer')
plt.show()
```

```python
manufacturer_condition_fair = clean_cars_df.groupby('manufacturer')['condition'].apply(lambda x: ((x == 'fair').sum() / (x != '').sum()) * 100).reset_index(name='Fair percentage')
manufacturer_condition_fair = manufacturer_condition_fair.sort_values('Fair percentage', ascending=True)
plt.barh(y = manufacturer_condition_fair['manufacturer'].values, width = manufacturer_condition_fair['Fair percentage'].values, color = 'brown')
plt.title('Percentage of fair condition cars per manufacturer')
plt.xlabel('Percentage of fair condition cars')
plt.ylabel('Manufacturer')
plt.show()
```

```python
manufacturer_condition_salvage = clean_cars_df.groupby('manufacturer')['condition'].apply(lambda x: ((x == 'salvage').sum() / (x != '').sum()) * 100).reset_index(name='Salvage percentage')
manufacturer_condition_salvage = manufacturer_condition_salvage.sort_values('Salvage percentage', ascending=True)
plt.barh(y = manufacturer_condition_salvage['manufacturer'].values, width = manufacturer_condition_salvage['Salvage percentage'].values, color = 'brown')
plt.title('Percentage of salvage cars per manufacturer')
plt.xlabel('Percentage of salvage cars')
plt.ylabel('Manufacturer')
plt.show()
```

```python
manufacturer_condition_nan = clean_cars_df.groupby('manufacturer')['condition'].apply(lambda x: ((pd.isna(x) == True).sum() / (x != '').sum()) * 100).reset_index(name='Unlisted percentage')
manufacturer_condition_nan = manufacturer_condition_nan.sort_values('Unlisted percentage', ascending=True)
plt.barh(y = manufacturer_condition_nan['manufacturer'].values, width = manufacturer_condition_nan['Unlisted percentage'].values, color = 'brown')
plt.title('Percentage of unlisted condition cars per manufacturer')
plt.xlabel('Percentage of unlisted condition cars')
plt.ylabel('Manufacturer')
plt.show()
```

Investigating price based on condition:

```python
condition_prices = clean_cars_df.groupby('condition')['price'].mean().reset_index(name='Average price')
condition_prices = condition_prices.sort_values('Average price', ascending=True)
plt.bar(x = condition_prices['condition'].values, height = condition_prices['Average price'].values, color = 'grey')
plt.title('Average price per listed car condition')
plt.xlabel('Average price')
plt.ylabel('Listed car condition')
plt.show()
```
