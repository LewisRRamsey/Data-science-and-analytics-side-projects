```python
used_cars_df.drop(['county', 'size'], axis = 1, inplace = True)
```

```python
used_cars_df.isnull().sum()
# The null values are going to be left in the dataset and excluded in each visual representation, as removing rows that contain a null value leads to almost every row being deleted.
```

```python
# finding price outliers

import numpy as np

# Filtering rows where price = 0
filtered_cars_df = used_cars_df[used_cars_df['price'] != 0]

car_prices = filtered_cars_df["price"].values
average_price = np.percentile(car_prices, 50)
print(f'Average price: {average_price}')
q3_price = np.percentile(car_prices, 75)
print(f'Upper quartile: {q3_price}')
q1_price = np.percentile(car_prices, 25)
print(f'Lower quartile: {q1_price}')
iqr = q3_price - q1_price
print(f'Interquartile range: {iqr}')
```
